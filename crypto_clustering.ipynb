# Clustering Crypto

```python
# Import Dependencies
import hvplot.pandas
import pandas as pd
from path import Path
import plotly.express as px
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

### Deliverable 1: Preprocessing the Data for PCA

```python
# Load the crypto_data.csv dataset.
# NOTE: Load user pre-processed version 'crypto_data_symbol_added_edited.csv'
# YOUR CODE HERE

file_path = Path('Data/crypto_data_symbol_added_edited.csv')

crypto_df = pd.read_csv(file_path, skiprows=0, index_col='Symbol')
```

```python
# Check DataFrame
print(crypto_df.shape)
crypto_df.head(10)
```

```python
# Check DataFrame
crypto_df.count()
```

```python
# Check DataFrame
crypto_df.dtypes
```

```python
# Keep all the cryptocurrencies that are being traded.
# YOUR CODE HERE
crypto_traded_df = crypto_df.loc[(crypto_df['IsTrading'] == True)]
```

```python
# Check DataFrame
print(crypto_traded_df.shape)
crypto_traded_df.head(10)
```

```python
# Keep all the cryptocurrencies that have a working algorithm.
# YOUR CODE HERE
crypto_traded_working_algorithm_df = crypto_traded_df.loc[(crypto_traded_df['Algorithm'] != "")]
```

```python
# Check DataFrame
print(crypto_traded_working_algorithm_df.shape)
crypto_traded_working_algorithm_df.head(10)
```

```python
# Remove the "IsTrading" column. 
# YOUR CODE HERE
crypto_traded_working_algorithm_df.drop(['IsTrading'], axis=1, inplace=True)
print(crypto_traded_working_algorithm_df.shape)
crypto_traded_working_algorithm_df.head(10)
```

```python
# Remove rows that have at least 1 null value.
# YOUR CODE HERE
crypto_traded_working_algorithm_df = crypto_traded_working_algorithm_df.dropna(how='any')
print(crypto_traded_working_algorithm_df.shape)
crypto_traded_working_algorithm_df.head(10)
```

```python
# Keep the rows where coins are mined ('TotalCoinsMined' > 0).
# YOUR CODE HERE
crypto_traded_working_algorithm_non_zero_coins_df = crypto_traded_working_algorithm_df.loc[(crypto_traded_working_algorithm_df['TotalCoinsMined'] > 0)]
print(crypto_traded_working_algorithm_non_zero_coins_df.shape)
crypto_traded_working_algorithm_non_zero_coins_df.head(10)
```

```python
# Create a new DataFrame that holds only the cryptocurrencies names.
# YOUR CODE HERE
coin_names_df = crypto_traded_working_algorithm_non_zero_coins_df.loc[:, ['CoinName']]
print(coin_names_df.shape)
#print(type(coin_names_df))
coin_names_df.head(5)
```

```python
# Drop the 'CoinName' column since it's not going to be used on the clustering algorithm.
# YOUR CODE HERE
crypto_traded_working_algorithm_non_zero_coins_no_names_df = crypto_traded_working_algorithm_non_zero_coins_df.drop(['CoinName'], axis=1)
print(crypto_traded_working_algorithm_non_zero_coins_no_names_df.shape)
crypto_traded_working_algorithm_non_zero_coins_no_names_df.head(10)
```

```python
# Use get_dummies() to create variables for text features.
# YOUR CODE HERE
X = pd.get_dummies(data=crypto_traded_working_algorithm_non_zero_coins_no_names_df, columns=['Algorithm', 'ProofType'])
print(X.shape)
X.head(10)
```

```python
# Standardize the data with StandardScaler().
# YOUR CODE HERE
X = StandardScaler().fit_transform(X)
X[:5]
```

### Deliverable 2: Reducing Data Dimensions Using PCA

```python
# Using PCA to reduce dimension to three principal components.
# YOUR CODE HERE
pca = PCA(n_components=3)
crypto_pca = pca.fit_transform(X)
crypto_pca
```

```python
# Create a DataFrame with the three principal components.
# YOUR CODE HERE
col_names = ["PC 1", "PC 2", "PC 3"]
crypto_pca_df = pd.DataFrame(crypto_pca, columns=col_names, index=crypto_traded_working_algorithm_non_zero_coins_no_names_df.index)

print(crypto_pca_df.shape)
crypto_pca_df.head(10)
```

### Deliverable 3: Clustering Crytocurrencies Using K-Means

#### Finding the Best Value for `k` Using the Elbow Curve

```python
# Create an elbow curve to find the best value for K.
# YOUR CODE HERE

# Inertia Calculation Set-Up
inertia = []
k = list(range(1, 11))

# Calculate inertia for the range of K values
for i in k:
    km = KMeans(n_clusters=i, random_state=0)
    km.fit(crypto_pca_df)
    inertia.append(km.inertia_)

# Plot Set-Up
elbow_data = {"k": k, "inertia": inertia}

elbow_df = pd.DataFrame(elbow_data)

elbow_df.hvplot.line(x="k", y="inertia", xticks=k, title="Elbow Curve")
```

Running K-Means with `k=4`

```python
# Initialize the K-Means model.
# YOUR CODE HERE
model = KMeans(n_clusters=4, random_state=0)
```

```python
# Fit the model
# YOUR CODE HERE
model.fit(crypto_pca_df)
```

```python
# Predict clusters
# YOUR CODE HERE
predictions = model.predict(crypto_pca_df)
predictions
```

```python
# Create a new DataFrame including predicted clusters and cryptocurrencies features.
# Concatentate the crypto_df and pcs_df DataFrames on the same columns.
# YOUR CODE HERE
clustered_df = pd.concat([crypto_traded_working_algorithm_non_zero_coins_no_names_df, crypto_pca_df], axis=1, sort=False)

#  Add a new column, "CoinName" to the clustered_df DataFrame that holds the names of the cryptocurrencies. 
# YOUR CODE HERE
clustered_df["CoinName"] = coin_names_df["CoinName"]

#  Add a new column, "Class" to the clustered_df DataFrame that holds the predictions.
# YOUR CODE HERE
clustered_df["Class"] = model.labels_

# Print the shape of the clustered_df
print(clustered_df.shape)
clustered_df.head(10)
```
